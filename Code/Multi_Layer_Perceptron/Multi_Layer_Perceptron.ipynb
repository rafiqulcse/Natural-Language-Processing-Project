# **Import Libraries**

```python
import os
import shutil

import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt

tf.get_logger().setLevel('ERROR')
```

```python
import pandas as pd
```

```python
github_url = "https://raw.githubusercontent.com/rafiqulcse/Natural-Language-Processing-Project/main/Dataset/NLP-project-dataset.csv"

df = pd.read_csv(github_url)
df.head()
```

# **Preprocess Dataset**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
```

```python
tfidf = TfidfVectorizer(analyzer = 'word', max_features = 150)
X = tfidf.fit_transform(df['clean_comment'])
X
```

```python
y = df['category']
X = X.toarray()
X
```

```python
X = pd.DataFrame(X)
X
```

```python
feature = tfidf.vocabulary_
col_names = []
```

```python
for key, value in feature.items():
    print(key, ' : ', value)
    col_names.append(key)
```

```python
X.columns = col_names
col_names
X
```

# **Training-Testing Split**

```python
X.shape, y.shape
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_train.shape
```

# **MLP Model Implementation**

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(3,2), max_iter=500, activation='relu')
```

```python
mlp.fit(X_train, y_train)
```

# **Testing**

```python
pred = mlp.predict(X_test)
pred
```

# **Model Evaluation**

```python
confusion_matrix(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```
